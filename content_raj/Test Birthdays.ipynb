```python
def get_bday(url):
    #url = 'http://en.wikipedia.org/wiki/Bill_Clinton'
    website_html = requests.get(url).text
    dom = web.Element(website_html)
    bday = dom.by_class('bday')[0].content
    return bday


 
```

```python
get_bday("http://en.wikipedia.org/wiki/Bill_Clinton")
```

```python
get_bday("http://en.wikipedia.org/wiki/Jawaharlal_Nehru")
```
